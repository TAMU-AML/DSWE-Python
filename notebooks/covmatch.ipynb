```python
from dswe import CovMatch
import numpy as np
import pandas as pd
```

### Load dataset and feature selection

```python
df1 = pd.read_csv('../../Inland_Offshore_Wind_Farm_Dataset1/Inland Wind Farm Dataset1(WT1).csv')
df2 = pd.read_csv('../../Inland_Offshore_Wind_Farm_Dataset1/Inland Wind Farm Dataset1(WT2).csv')

cols = [1,2]
circ_pos = 1

Xlist = [df1.iloc[:100, cols].to_numpy(), df2.iloc[:500, cols].to_numpy()]
ylist = [df1.iloc[:100, 6].to_numpy(), df2.iloc[:500, 6].to_numpy()]
```

### Fit the CovMatch model

```python
covmatch = CovMatch(Xlist, ylist, circ_pos)
```
